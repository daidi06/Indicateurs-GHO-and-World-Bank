```python
#Return as a dataFrame
source = pd.DataFrame(wb.source.list())

def IndicatorsMetaDataWB(IndicatorId):
    for i in list(source['id']):
        wb.db = i
        for k,v in wb.search(f"'{IndicatorId}'").metadata.items():
            for elem in v:
                if elem.id == IndicatorId:
                    metaD = wb.series.metadata.get(IndicatorId)
        break 
    return(metaD)
```

```python
IndicatorsMetaDataWB('SP.DYN.LE00.IN')
```

```python
def IndicatorsDataWB(IndicatorId):
    data=wb.data.DataFrame(IndicatorId, 
                             wb.region.members('AFR'), 
                             #time = range(2000, 2023,1), 
                             skipBlanks=True, 
                             columns='series', 
                             labels = True)
    data.reset_index()
    return(data)
```
